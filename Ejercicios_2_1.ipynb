### 2.1.1 Encuentre todos los puntos fijos del flujo.

```
import matplotlib.style
matplotlib.style.use('classic')
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
%matplotlib inline
import matplotlib.pyplot as plt
```

En un punto fijo, el flujo debe ser cero.

si tomamos a x=0, entonces sin(x) sera cero.

tomamos a X=n*pi y n debe pertenecer a los numeros naturales. Entonces la funcion tendra infinitos puntos fijos.


```
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x,y_sin)
plt.title("seno")
plt.legend(["seno"])
```

### 2.1.2 ¿En qué puntos x el flujo tiene mayor velocidad hacia la derecha?

La velocidad y su dirección están determinadas por el valor de x. Entonces, el máximo valor positivo de la función es la mayor velocidad a la derecha.

sin(x)=1 

x= pi/2 +n*2*pi en donde n pertenece a todos los naturales N.

El flujo tiene la mayor velocidad hacia la derecha en todos los valores x

```
n=0.3 ;
x = np.arange(0,  np.pi/2 + n*2*np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x,y_sin)
plt.title("seno")
plt.legend(["seno"])
```

### 2.1.3
a) Encuentra la aceleración del flujo x en función de x.

ya tenemos la función en su forma de velocidad,para encontrar la aceleracion derivamos la funcion y obtenemos:

\begin{equation}
\ddot { x } =cos {x}*sin{x}
\end{equation}

```
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)*np.sin(x)

plt.plot(x,y_cos)
plt.title("cos")
plt.legend(["cos"])
```

### b) Encuentre los puntos donde el flujo tiene máxima aceleración positiva.

La aceleración y su dirección están determinadas por el valor de x. Entonces, el máximo valor positivo de la función es la mayor aceleracion positiva a la derecha.

cos(x)*sin(x)=1 

x= n*pi + 2*pi en donde n pertenece a todos los naturales N.

El flujo tiene la mayor aceleración hacia la derecha en todos los valores x

```
n=2 ;
x = np.arange(0,  n*np.pi + 2*np.pi, 0.1)
y_cos = np.cos(x)*np.sin(x)

plt.plot(x,y_cos)
plt.title("cos")
plt.legend(["cos"])
```
